```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("insurance_pre.csv")
print(dataset)
```

```python
dataset= pd.get_dummies(dataset,drop_first=True)
print(dataset)
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes' ]]
print(independent)
```

```python
dependent=dataset[["charges"]] 
print(dependent)
```

```python
from sklearn.model_selection import GridSearchCV  # grid means various para check it ,cv means various type split it
from sklearn.svm import SVR

param_grid = {'kernel':('linear', 'rbf','ploy','sigmoid'), 'C':[1, 10,100,1000],'gamma':['auto','scale']}

# ALL para in single line and also using key and value(dictionry)

grid=GridSearchCV(SVR(),param_grid ,refit=True,verbose=3,n_jobs=-1)#refit last final value is best model get
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
print(table)
```

```python
age_input= float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("children:"))
sex_male_input=int(input("sex male 0 or 1:"))
smoker_yes_input=int(input("smoker 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
