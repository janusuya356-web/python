# univaritae analaysis


```python
import seaborn as sb
sb.distplot(dataset['ssc_p'],hist=True,kde=True,kde_kws={'color':'red'})
```

# Bivariate analaysis: Distribution plot

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset)
plt.show()
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='hex')
plt.show()
```

```python
sb.jointplot(x='ssc_p',y='salary',data=dataset,kind='kde')
plt.show()
```

# pairplot

```python
sb.pairplot(dataset,hue='gender',diag_kind="kde",kind="scatterplot",palette="husl")
plt.show()
```

# Categorical plot

```python
sb.stripplot(x='ssc_b',y='ssc_p',data=dataset)
plt.show()
```

```python
sb.stripplot(x='ssc_b',y='ssc_p',data=dataset,jitter=True)
plt.show()
```

```python
sb.swarmplot(x='gender',y='ssc_p',data=dataset)
plt.show()
```
