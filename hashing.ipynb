```python
import pandas as pd
import numpy as np
import itertools
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('news.csv',index_col=None)
df
```

```python
dataset=df.drop("subject",axis=1)
```

```python
dataset
```

```python
y=dataset["LABEL"]
```

```python
y
```

```python
x=dataset["text"]
```

```python
x
```

```python
X_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=53)
```

```python
hash_vec=HashingVectorizer(stop_words='english',n_features=2**10,alternate_sign=False) #common used words remove it
X_train_hash=hash_vec.transform(X_train)#preprocessing/ x_train(input)/model create and predict
print(X_train_hash)
x_test_hash=hash_vec.transform(x_test) # change to the vector
```

```python
model=MultinomialNB()

model.fit(X_train_hash,y_train) #model create/ y_train(output)
pred=model.predict(x_test_hash) # predict
score=metrics.accuracy_score(y_test,pred)
print("accuracy: %0.3f " % score)

cm=metrics.confusion_matrix(y_test,pred,labels=['FAKE','REAL'])
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,pred)
```
