```python
def oddeven():
    num=int(input("enter a number"))
    if(num%2==0):
            print("even number")
            msg="even"
    else:
            print("odd  number")
            msg="odd"
    return msg
oddeven()   
```

```python
def Elegible():
        gender = input("Enter gender:")
        age = int(input("Enter age: "))

        if gender == "male" and age >= 21:
            print("Eligible for marriage")
        elif gender == "female" and age >= 18:
            print("Eligible for marriage")
        else:
            print("Not eligible for marriage")
Elegible()
```
