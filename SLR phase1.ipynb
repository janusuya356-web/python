```python
import pandas as pd# pandas is a libary to create a table
dataset=pd.read_csv("Salary_Data.csv")# dataset to assinging a pandas to using read_csv variable to include a Salary_data.csv
print(dataset)
```

```python
independent=dataset[["YearsExperience"]]#independent is buying a input
print(independent)
```

```python
dependent=dataset[["Salary"]] #The dependent is a buying a output
print(dependent)
```

```python
from sklearn.model_selection import train_test_split# this part to be split the train and test because of after finish this part then going do the model

x_train,x_test,y_train,y_test = train_test_split(independent, dependent, test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression# It is used to find a slope and bias
regressor=LinearRegression()# weight and bias

regressor.fit(x_train,y_train)# we using the train means to the weight
 
LinearRegression()
weight=regressor.coef_# the coef means slope of the variable
print(weight)
bais=regressor.intercept_ # the intercept means origin
print(bais)
```

```python
y_pred=regressor.predict(x_test) # in these process going to evaluvate
from sklearn.metrics import r2_score# r_score by using to find a postive or negative
r_score=r2_score(y_test,y_pred)
print(r_score)
```
