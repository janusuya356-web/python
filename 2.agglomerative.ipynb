```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_customer.csv")
print(dataset)
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
from  sklearn.cluster import AgglomerativeClustering
clusmodel= AgglomerativeClustering(n_clusters=5)
label=clusmodel.fit_predict(x)
label
```

```python
supervised=pd.DataFrame(dataset) 
```

```python
supervised['cluster_group']=label# coloumn add it
supervised
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4]  #hue color,fit_reg regression is no se we put flase
                 ,hue=supervised.columns[5],fit_reg=False,legend=True,legend_out=True)
```
